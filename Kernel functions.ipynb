## A demonstration of $k(x,x') = x^{T} x$

First lets consider the plot of the covaraince function. To make things easy, lets consider the 2D case where we only have 2 samples in our Gaussian process, which we know i.e. $x_1 = 1$ and $x_2 = 2$.

Then the covariance matrix becomes:

$$ k(x,x') =\begin{bmatrix}
    1 & 2\\
    2 & 4 \\
    \end{bmatrix}$$
    
Plotting the multivaraite normal with 0 mean and the above covariance matrix, we get:


```python
import numpy as np
import matplotlib.pyplot as plt

means = [0,0]
covariance = np.array([[1,2],[2,4]])

samples = np.random.multivariate_normal(means, covariance, 100)

plt.scatter(samples[:,0], samples[:,1])
plt.show()
```

We see that if $x_1 = 1$, then $x_2 = 2$. i.e. the value of $x_2$ is double that of $x_1$. Taking another sample of this line at $x_1 = 1.5$ gives us $x_2 = 3$, the line at $x_1 = 2$ gives us $x_2 = 4$ and $x_1 = 2.5$ gives $x_2 = 5$. The plot for these lines can be seen below:

```python
xs = np.arange(1,2.05,0.05)
print(xs)
ys = xs
ys2 = xs * 1.5
ys3 = xs * 2
ys4 = xs * 2.5
plt.plot(xs, ys) ; plt.xlim([0, 3]) ; plt.ylim([0, 4])
plt.plot(xs, ys2) ; plt.xlim([0, 3]) ; plt.ylim([0, 4])
plt.plot(xs, ys3) ; plt.xlim([0, 3]) ; plt.ylim([0, 4])
plt.plot(xs, ys4) ; plt.xlim([0, 3]) ; plt.ylim([0, 4])
plt.show()
```

We are simply taking samples of the multivariate normal distribution given above. Lets go a step further and consider this in the third dimension,  with a point $x_3 = 3$.

```python
means = [0,0,0]
covariance = np.array([[1,2,3],[2,4,6],[3,6,9]])

samples = np.random.multivariate_normal(means, covariance, 200)
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(samples[:,0], samples[:,1], samples[:,2], color = "green")
ax.view_init(5, 70)
plt.title("A plot of mvt norm with covariance k(x,x') = x^Tx' (3 dimensions)")
 
# show plot
plt.show()
```

Lets now take samples from this plot to consider a gaussian process with 3 samples at 1,2,3. 

```python
xs = np.arange(1,3.05,0.05)

ys = xs
ys2 = xs * 1.5
ys3 = xs * 2
ys4 = xs * 2.5
plt.plot(xs, ys) ; plt.xlim([0, 4]) ; plt.ylim([0, 6])
plt.plot(xs, ys2) ; plt.xlim([0, 4]) ; plt.ylim([0, 6])
plt.plot(xs, ys3) ; plt.xlim([0, 4]) ; plt.ylim([0, 6])
plt.plot(xs, ys4) ; plt.xlim([0, 4]) ; plt.ylim([0, 6])
plt.show()
```

We see the same graphs as above, which are now extended out even further to 3. 

```python
x = np.arange(-20,20,0.1)
y = np.arange(-20,20,0.1)

X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5*((X/5)+(2*Y/5))**2)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)
ax.view_init(20, 0)
plt.show()
```
